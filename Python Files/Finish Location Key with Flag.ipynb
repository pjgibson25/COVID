# Read JHU data

```python
import os
import numpy as np
import pandas as pd
import datetime

Location_key = pd.read_csv('../FIPS_tags/locationkey_without_flag.csv')
data = pd.read_csv('../ALL_FIPS_datetime_cases.csv')

present_data = np.array(data.FIPS)
all_fips = np.array(Location_key.FIPS)

cond = [all_fips[i] in present_data for i in np.arange(0,len(all_fips))]
flag = np.array(cond).astype(int)
flag_pd = pd.DataFrame(flag, columns=['Flag'])

final_locations = pd.concat((Location_key,flag_pd), axis=1)
final_locations.to_csv('../Location_key.csv',index=False)


```
